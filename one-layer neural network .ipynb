```python
data=[[0.08,0.72,1.0],
      [0.01,1.00,0.0],
      [0.26,0.58,1.0],
      [0.35,0.95,0.0],
      [0.45,0.15,1.0],
      [0.60,0.30,1.0],
      [0.70,0.65,0.0],
      [0.92,0.45,0.0]]

wei=[1.00, -1.00]
bias = 0.20
dataf = pd.DataFrame(data, columns=["data1", "data2", "target"])
```

```python
import numpy as np
import pandas as pd

def sigmoid(x):
    return 1/(1 + np.exp(-(x)))
     
def loss(pred, target):
    return (pred - target)**2
    
def derivative_Sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))
    
def derivative_loss(pred, target):
    return 2*(pred - target)

def predict(data, weights, threshold = 0.5, bias = 0.20):
    pred = sum(np.array(data) * np.array(weights))
    #pred = sigmoid(pred)
    return 1.0 if pred + bias >= threshold else 0.0

def accuracy(data, target, weights, threshold = 0.0):
    correct_pred=0.0
    preds = []

    for i in range(len(data)):
        pred=predict(data.iloc[i][:], weights)
        preds.append(pred)
        
        if pred==target[i]:
            correct_pred += 1.0
    
    print("predictions: " + str(preds))
    print("targets: " + str(list(target)))
    
    return correct_pred/len(data)

def train(data, target, weights, epochs=10, stop=True, alpfa = 0.001):
    
    for epoch in range(epochs):
        print("\n")
        print("Epoch: " + str(epoch))
        current_acc = accuracy(data, target, weights)
        print("accuracy: " + str(current_acc))
        
        if current_acc == 1 and stop:
            print("accuracy equals to 1 - training stoped")
            break
        
        lossTotal = []
        for i in range(len(data)):
            dl = float(derivative_loss(predict(data.iloc[i][:], weights), float(target.iloc[i])))
            lossTotal.append(dl)
        print("loss: " + str(lossTotal))
                                                             
        for j in range (len(weights)):
            gradient_ = np.mean((np.dot(data.iloc[:, j], lossTotal)))    
            
            print("gradient: " + str(gradient_))
            weights[j] = weights[j] - alpfa * gradient_
            
        print("weights: " + str(weights))
                
    return weights, current_acc
```

# dla alfa = 0.001

```python
data_feature = dataf.drop(columns = 'target')
data_class = dataf['target']

a, b = train(data_feature, data_class, wei.copy(), epochs=1000, alpfa = 0.001)
```
